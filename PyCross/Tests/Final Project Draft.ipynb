```python
def pixelate_bin(img, window, threshold):
    n, m = img.shape[:2]
    n, m = n - n % window, m - m % window
    img1 = np.zeros((n,m))
    for x in range(0, n, window):
        for y in range(0, m, window):
            if img[x:x+window,y:y+window].mean() > threshold:
                img1[x:x+window,y:y+window] = 1
    return img1

img = plt.imread('galaxy.png')

plt.imshow(pixelate_bin(img, 5, .2), cmap="Greys")
plt.colorbar()
```

```python
class CellValue():
    WHITE = 0
    BLACK = 1

def _generate_hints(row):
    clues = []

    next_idx = 0
    while next_idx < len(row):
        try:
            next_black_index = row[next_idx:].index(CellValue.BLACK) + next_idx

            try:
                next_white_index = row[next_black_index:].index(CellValue.WHITE)
                clues.append(next_white_index)
                next_idx = next_black_index + next_white_index

            except ValueError:
                remaining_length = len(row[next_black_index:])
                clues.append(remaining_length)

                break
        except ValueError:
            next_idx = len(row)

    return clues
```
