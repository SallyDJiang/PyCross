```python
def track_points(point, action, img):
    if action == "add": 
        point += 1
    if action == "sub": 
        point -= 1
    ax.text(x = np.shape(img)[0], y = np.shape(img)[1]/2, s="Score: "+str(point), fontsize=15)
    return point
```

```python
# subtract points
points = 0
fig, ax = plt.subplots()

ax.imshow(img)
points = track_points(points, "sub", img)
```
